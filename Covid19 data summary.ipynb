```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('owid-covid-data.csv')

print(df.columns)
print(df.head())
print(df.isnull().sum())

# Step 3:
countries = ['Kenya', 'United States', 'China']
df = df[df['location'].isin(countries)]

df = df.dropna(subset=['date', 'total_cases'])

df['date'] = pd.to_datetime(df['date'])

df.fillna(method='ffill', inplace=True)

# Step 4:
plt.figure(figsize=(12, 6))
for country in countries:
    country_df = df[df['location'] == country]
    plt.plot(country_df['date'], country_df['total_cases'], label=country)
plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 5:
plt.figure(figsize=(12, 6))
for country in countries:
    country_df = df[df['location'] == country]
    plt.plot(country_df['date'], country_df['total_vaccinations'], label=country)
plt.title('Total Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# Step 7: Summary
# - China has the highest number of total cases among the selected countries.
# - The USA shows the fastest initial vaccination rollout.
# - Kenyas case growth is significantly lower in comparison.

```
